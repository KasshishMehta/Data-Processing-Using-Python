```python
import pandas as pd
import matplotlib.pyplot as plt
```

Metrics Used:

CTR (Link Click-Through Rate): Measures the effectiveness of the ad in driving traffic to the link. A high CTR indicates that the ad is compelling and engaging to the audience.

CPC (Cost per Link Click): Measures the cost efficiency of generating clicks. A low CPC indicates that the ad is cost-effective in driving clicks.

Purchase ROAS (Return on Ad Spend): Measures the efficiency and profitability of the ad spend. A high ROAS indicates that the ad is generating significant revenue relative to its cost.


```python
ads_data = pd.read_csv('/content/Sample ads data - Performance data.csv')

corrected_cols = {
    'Amount spent (INR)': 'Amount Spent (INR)',
    'Cost per result': 'Cost per Result'
}
ads_data.rename(columns=corrected_cols, inplace=True)

cols_to_analyze = [
    'Ad name', 'Gender', 'Age', 'Campaign name', 'Ad Set Name', 'Ad set ID',
    'CPM (cost per 1,000 impressions)', 'Results', 'Reach', 'Impressions', 'Amount Spent (INR)',
    'Cost per Result', 'CPC (cost per link click)', 'CTR (link click-through rate)', 'Purchase ROAS (return on ad spend)'
]

ads_data_filtered = ads_data.loc[:, cols_to_analyze]
ads_data_filtered.loc[:, 'Ad info'] = ads_data_filtered.apply(lambda row: f"{row['Ad name']} - {row['Gender']} - {row['Age']}", axis=1)

metrics = ['CTR (link click-through rate)', 'CPC (cost per link click)', 'Purchase ROAS (return on ad spend)']
ads_data_filtered = ads_data_filtered.dropna(subset=metrics)

ads_data_filtered.head()
```

```python
plt.figure(figsize=(15, 5))

for i, metric in enumerate(metrics, 1):
    plt.subplot(1, 3, i)
    plt.boxplot(ads_data_filtered[metric])
    plt.title(metric)
    plt.ylabel(metric)

plt.tight_layout()
plt.show()
```

```python
def detect_outliers(data, metric):
    Q1 = data[metric].quantile(0.25)
    Q3 = data[metric].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[metric] < lower_bound) | (data[metric] > upper_bound)]
    return outliers

outliers_ctr = detect_outliers(ads_data_filtered, 'CTR (link click-through rate)')
outliers_cpc = detect_outliers(ads_data_filtered, 'CPC (cost per link click)')
outliers_roas = detect_outliers(ads_data_filtered, 'Purchase ROAS (return on ad spend)')

outliers_combined = pd.concat([outliers_ctr, outliers_cpc, outliers_roas]).drop_duplicates()
outliers_combined.head()
```

Explanation:
Boxplot Visualization: Boxplots provide a visual summary of the data distribution and help in identifying outliers beyond the whiskers.
IQR Method: The interquartile range (IQR) method is used to detect outliers. Data points below Q1 - 1.5IQR or above Q3 + 1.5IQR are considered outliers.
Summary:
CTR, CPC, and ROAS are the key metrics used for performance analysis.
Boxplots provide a visual means to identify outliers.
IQR Method helps in systematically detecting outliers for each metric.
Recommendations can be made based on the identified outliers, focusing on scaling up high-performing ads and optimizing underperforming ones.

Methodology
Identify Overperforming Ads: Ads with high CTR, low CPC, and high ROAS.
Identify Underperforming Ads: Ads with low CTR, high CPC, and low ROAS.


```python
overperforming_ads = outliers_combined[
    (outliers_combined['CTR (link click-through rate)'] > ads_data_filtered['CTR (link click-through rate)'].median()) &
    (outliers_combined['CPC (cost per link click)'] < ads_data_filtered['CPC (cost per link click)'].median()) &
    (outliers_combined['Purchase ROAS (return on ad spend)'] > ads_data_filtered['Purchase ROAS (return on ad spend)'].median())
]

underperforming_ads = outliers_combined[
    (outliers_combined['CTR (link click-through rate)'] < ads_data_filtered['CTR (link click-through rate)'].median()) &
    (outliers_combined['CPC (cost per link click)'] > ads_data_filtered['CPC (cost per link click)'].median()) &
    (outliers_combined['Purchase ROAS (return on ad spend)'] < ads_data_filtered['Purchase ROAS (return on ad spend)'].median())
]
```

```python
overperforming_ads_display = overperforming_ads[['Ad info', 'CTR (link click-through rate)', 'CPC (cost per link click)', 'Purchase ROAS (return on ad spend)']]
underperforming_ads_display = underperforming_ads[['Ad info', 'CTR (link click-through rate)', 'CPC (cost per link click)', 'Purchase ROAS (return on ad spend)']]
```

```python
def plot_ads(data, metric, title, median_value):
    plt.figure(figsize=(10, 6))
    plt.bar(data['Ad info'], data[metric])
    plt.axhline(y=median_value, color='r', linestyle='--', label='Median')
    plt.title(title)
    plt.xlabel('Ad info')
    plt.ylabel(metric)
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()

for metric in metrics:
    median_value = ads_data_filtered[metric].median()
    plot_ads(overperforming_ads_display, metric, f'Overperforming Ads - {metric}', median_value)

for metric in metrics:
    median_value = ads_data_filtered[metric].median()
    plot_ads(underperforming_ads_display, metric, f'Underperforming Ads - {metric}', median_value)
```

```python
ads_combined = pd.merge(ads_data_filtered, ads_type_data, on='Ad name')

cols_to_analyze = [
    'Ad name', 'Gender', 'Age', 'Campaign name', 'Ad Set Name', 'Ad set ID',
    'CPM (cost per 1,000 impressions)', 'Results', 'Reach', 'Impressions', 'Amount Spent (INR)',
    'Cost per Result', 'CPC (cost per link click)', 'CTR (link click-through rate)', 'Purchase ROAS (return on ad spend)',
    'Media type'  ]

ads_combined_filtered = ads_combined.loc[:, cols_to_analyze]

ads_combined_filtered.loc[:, 'Ad info'] = ads_combined_filtered.apply(lambda row: f"{row['Ad name']} - {row['Gender']} - {row['Age']}", axis=1)

ads_combined_filtered.loc[:, 'Conversion Rate'] = ads_combined_filtered['Results'] / ads_combined_filtered['Impressions']
ads_combined_filtered.loc[:, 'ROI'] = ads_combined_filtered['Purchase ROAS (return on ad spend)'] * ads_combined_filtered['Amount Spent (INR)']
```

```python
media_type_performance = ads_combined_filtered.groupby('Media type').agg({
    'CTR (link click-through rate)': 'mean',
    'Conversion Rate': 'mean',
    'CPC (cost per link click)': 'mean',
    'ROI': 'mean'
}).reset_index()

media_type_performance.plot(x='Media type', kind='bar', figsize=(15, 8), subplots=True, layout=(2, 2), sharex=False, legend=False, title='Performance by Media Type')
plt.tight_layout()
plt.show()
```

```python
gender_type_performance = ads_combined_filtered.groupby('Gender').agg({
    'CTR (link click-through rate)': 'mean',
    'Conversion Rate': 'mean',
    'CPC (cost per link click)': 'mean',
    'ROI': 'mean'
}).reset_index()

# Plot performance by gender type
gender_type_performance.plot(x='Gender', kind='bar', figsize=(15, 8), subplots=True, layout=(2, 2), sharex=False, legend=False, title='Performance by Media Type')
plt.tight_layout()
plt.show()
print(gender_type_performance)
```

```python
age_type_performance = ads_combined_filtered.groupby('Age').agg({
    'CTR (link click-through rate)': 'mean',
    'Conversion Rate': 'mean',
    'CPC (cost per link click)': 'mean',
    'ROI': 'mean'
}).reset_index()

# Plot performance by age type
age_type_performance.plot(x='Age', kind='bar', figsize=(15, 8), subplots=True, layout=(2, 2), sharex=False, legend=False, title='Performance by Media Type')
plt.tight_layout()
plt.show()
print(age_type_performance)
```
